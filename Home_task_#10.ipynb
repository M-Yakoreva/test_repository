## Задание 1
### Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
tables_ = ratings.merge(movies, on='movieId', how='left')#объединяем 2 датасета
tables_.head()
```

```python
max_= ratings['rating'].max()
filter_max = tables_[(tables_ ['rating'] == max_)]#фильтруем датасет по max оценке'5'
film_raiting_max = filter_max.groupby(['movieId' , 'title'])['movieId'].count()#считаем количество оценок по каждому фильму
film_raiting_max.sort_values(ascending = False).head(5)#сортируем оценки от большего к меньшему
```
