```python
# 1. По показателю surgery? мы имеем 1 пропуск, это категория, которая отражает была ли сделана операция лошади или нет. 
# Так как есть только 2 варианта ответа, пропуск заполняем наиболее часто встречающимся (Модой).

# 2. По показателю rectal temperature 20% пропусков (60). Проанализировав средние показатели (ср. арифметическое, моду, медиану),
# max и min значение, а также отсутсвие выбросов, можно принять решение о замене пропусков средним арифметическим значением.

# 3.По показателю pulse 8% пропусков (24). Есть существенные выбросы в данных.
# Нормальный пульс у лошади в состоянии покоя составляет от 25 до 40 ударов в минуту. 
#Если пульс измеряли после тренировки лошади или в ожидании предстоящей тренировки у большинства лошадей пульс поднимается до 48-60 уд/мин.
#Если лошадь напугана или находится в состоянии шока симпатическая нервная система выделяет эпинефрин (адреналин),
#который пересиливает тонус блуждающего нерва и может быстро поднять частоту пульса лошади до 220-240 уд/мин. Возможно выбросы связаны именно с этим обстоятельством. 
# Согласно приведенным в таблице данным на пульс не влияет факт проведения операции на пульс не влияет факт проведения операции. 
# Заменим все пропуски Модой.

# 4. По показателю respiratory rate 19% пропусков (58).Все рассчитанные средние значения значительно превышают нормальную частоту дыхания,
#которая должны быть от 8-10. Это скорее вего связано с болезнью животных.
#Частота дыхания возрастает при стрессе, повышении температуры и респираторных заболеваниях.
# Так как есть выбросы в данных, заменим существующие пропуски медианой.

# 5. По показателю temperature of extremities 18.7% пропусков (56). Есть выбросы. Значения Моды и Медианы равны.
# Заменим пропуски Модой, так как это категориальый показатель.

# 6. По показателю pain 18.4% пропусков (55).Есть выбросы. Значения Моды и Медианы равны. 
# Заменим пропуски Модой, так как это категориальый показатель.

# 7.По показателю outcome мы имеем 1 пропуск, это категория. Есть выбросы. Значения Моды и Медианы равны. 
# Заменим пропуски Модой,так как это категориальый показатель.
```

```python
df_new['surgery?'].fillna(df_new['surgery?'].round().mode()[0], inplace=True)
df_new['surgery?'].fillna(df_new['surgery?'].round().mode()[0]).isna().sum()
df_new['rectal temperature'].fillna(df_new['rectal temperature'].mean, inplace=True)
df_new['pulse'].fillna(df_new['pulse'].round().mode()[0], inplace=True)
df_new['respiratory rate'].fillna(df_new['respiratory rate'].median, inplace=True)
df_new['temperature of extremities'].fillna(df_new['temperature of extremities'].mode()[0], inplace=True)
df_new['pain'].fillna(df_new['pain'].round().mode()[0], inplace=True)
df_new['outcome'].fillna(df_new['outcome'].round().mode()[0], inplace=True)
```
